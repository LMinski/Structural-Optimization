```python
import numpy as np
import pandas as pd
import random
```

```python
class Viga:
    def __init__(self, forca, comprimento, modulo, momento, tipo):
        self.forca = forca
        self.comprimento = comprimento
        self.modulo = modulo
        self.momento = momento
        self.viga = tipo
    
    def calc_deflexao_maxima(self):
        if self.viga == "biapoiada":
            return (self.forca * self.comprimento ** 3) / (48 * self.modulo * self.momento)
        elif self.viga == "engastada":
            return (self.forca * self.comprimento ** 3) / (3 * np.sqrt(3) * self.modulo * self.momento)
        elif self.viga == "simplesmente apoiada":
            return (5 * self.forca * self.comprimento ** 4) / (384 * self.modulo * self.momento)
        else:
            raise ValueError("Tipo de viga inválido.")
    
    def calc_deflexao_ponto(self, x):
        if self.viga == "biapoiada":
            return (self.forca * x ** 2 * (3 * self.comprimento - x)) / (6 * self.modulo * self.momento * self.comprimento)
        elif self.viga == "engastada":
            return (self.forca * (self.comprimento - x) * x ** 2 * (3 * self.comprimento - x)) / (6 * np.sqrt(3) * self.modulo * self.momento * self.comprimento)
        elif self.viga == "simplesmente apoiada":
            return (self.forca * x ** 2 * (self.comprimento ** 2 - 4 * self.comprimento * x + 3 * x ** 2)) / (8 * self.modulo * self.momento * self.comprimento ** 3)
        else:
            raise ValueError("Tipo de viga inválido.")
```

```python
results = pd.DataFrame(columns=["Força", "Comprimento", "Módulo de Elasticidade", "Momento de Inércia", "Tipo de Viga", "Deflexão Máxima"])
a = random.randint(1,10)
print(a)
tipos = ['biapoiada','engastada','simplesmente apoiada']
moduloelast= [200000000000, 80000000000]
```

```python
# Exemplo de uso da classe para uma viga
for i in range(0,1000):
    for i in range(1001):

        forca = random.randint(1000,150000) # em Newtons
        comprimento = random.randint(1,15) # em metros
        modulo = moduloelast[random.randint(0,1)] # em pascal
        momento = 8.33*10**-5 # em metros^4
        viga = tipos[random.randint(0,2)]
    viga = Viga(forca, comprimento, modulo, momento, viga)

    deflexao_maxima = viga.calc_deflexao_maxima()
    deflexao_maxima = np.round(deflexao_maxima*1000, 2)
    row = [forca, comprimento, modulo, momento, viga, deflexao_maxima]
    results.loc[len(results)] = row
    results.tail()
```
