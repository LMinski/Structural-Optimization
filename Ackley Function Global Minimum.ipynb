```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#função
def ackley(x):
    sum = -20. * np.exp(-0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))) - np.exp(0.5 *(np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))) + 20. + np.e
    return sum
```

```python
 #plotagem

x1_v = np.linspace(-5,5,500)  
x2_v = np.linspace(-5,5,500)   
x1_arr, x2_arr = np.meshgrid(x1_v,x2_v)

f_arr=[[ackley([x1_,x2_]) for x1_ in x1_v] for x2_ in x2_v]
f_arr=np.array(f_arr)

ax=plt.figure().add_subplot(projection='3d')
surf=ax.plot_surface(x1_arr,x2_arr,f_arr)
plt.figure()
plt.contour(x1_arr,x2_arr,f_arr, 25)
plt.plot(-2.17811931e-09,-4.85151194e-09,'r*')
plt.show()
```

```python
#suposições dos pontos, utilização dos algoritmos e plotagem da posição x e valor da função no ponto x1 e x2
from urllib.response import addinfo
from scipy.optimize import differential_evolution
from scipy.optimize import shgo
from scipy.optimize import dual_annealing
from scipy.optimize import direct
bounds = [(-5,5), (-5,5)]
print('Differential Evolution')
result1 = differential_evolution(ackley,bounds, maxiter=100, popsize=10)
print(result1)
print('')
print('shgo')
result2 = shgo(ackley,bounds,n=100, iters=10)
print(result2)
print('')
print('Dual Annealing')
result3 = dual_annealing(ackley,bounds,maxiter=100)
print(result3)
print('')
print('Direct')
result4 = direct(ackley,bounds,maxiter=100)
print(result4)
```
